```python
videos=[]
channels=[]
titles=[]
descriptions=[]
channeltitle=[]
```

```python
for i in response["items"]:
  for j in i:
    dict1=i[j]
    for k in dict1:
      if k=='videoId':
        videos.append(dict1[k])
      if k=='channelId':
        channels.append(dict1[k])
      if k=='title':
        titles.append(dict1[k])
      if k=='description':
        descriptions.append(dict1[k])
      if k=='channelTitle':
        channeltitle.append(dict1[k])
```

##Caption Generation

```python
errorlist=[]
finalcaptionlist=[]
b=0
for o in videos:
  try:
    cap=YouTubeTranscriptApi.get_transcripts([o])
  except:
    errorlist.append(b)
    b=b+1
    continue
  finalcaption=""
  for i in cap:
    my_dict={}
    my_dict=i
    for j in my_dict:
      my_list=my_dict[j]
      for k in my_list:
        my_dict1=k
        for m in my_dict1:
          if m=='text':
            finalcaption=finalcaption+my_dict1[m]+" "
  finalcaptionlist.append(finalcaption)  
  b=b+1
```

```python
for i in videos:
  print(i)
```

```python
j=0
for i in errorlist:
  videos.pop(i-j)
  channels.pop(i-j)
  descriptions.pop(i-j)
  titles.pop(i-j)
  channeltitle.pop(i-j)
  j=j+1
```

```python
for i in finalcaptionlist:
  print(i)
```

```python
for i in videos:
  print(i)
```

```python
finalist=[]
for i in range(len(videos)):
  listing=[]
  listing.append(i+1)
  listing.append(videos[i])
  listing.append(channels[i])
  listing.append(titles[i])
  listing.append(descriptions[i])
  listing.append(channeltitle[i])
  listing.append(finalcaptionlist[i])
  finalist.append(listing)
```

```python
for i in finalist:
  print(i)
```

```python
fields = ['Serial No','Video Id', 'Channel Id', 'Titles', 'Description', 'Channel Title','Captions'] 
with open('/content/drive/MyDrive/Hate Speech Dataset/Racism Dataset.csv', 'w') as f:
    write = csv.writer(f)
    write.writerow(fields)
    write.writerows(finalist)
```
